```python
import pandas as pd
```

```python
#(A) Selecting authors last names in descending order.
pd.read_sql('SELECT last FROM authors ORDER BY last DESC', 
            connection)
```

```python
#(B) Selecting book titles in ascending order.
pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
           connection)
```

```python
#(C) Selecting all books for a specific author.
pd.read_sql('''SELECT first, last, title, copyright, titles.isbn FROM authors
            INNER JOIN author_ISBN ON authors.id = author_ISBN.id
            INNER JOIN titles ON author_ISBN.isbn = titles.isbn
            ORDER BY title ASC''', connection)
```

```python
#(D) Inserting a new author into the authors table.
#Creating a cursor.
cursor = connection.cursor()
```

```python
#Inserting the author.
cursor = cursor.execute('''INSERT INTO authors (first, last)
                        VALUES ('Karolina', 'Smoter')''')
```

```python
#Printing.
pd.read_sql('SELECT first, last FROM authors',
           connection)
```

```python
#(E) Inserting a new title for an author.
#Into the author_isbn table.
cursor = cursor.execute('''INSERT INTO author_ISBN (id, isbn)
                        VALUES ('9', '0000000000')''')
```

```python
#Into the titles table.
cursor = cursor.execute('''INSERT INTO titles (isbn, title, edition, copyright)
                        VALUES ('0000000000', 'New Added Book', 1, '2022')''')
```

```python
#Printing.
pd.read_sql('SELECT * FROM titles',
           connection)
```

```python
#Printing.
pd.read_sql('SELECT * FROM author_ISBN',
           connection)
```

```python
#Closing the database.
connection.close()
```
